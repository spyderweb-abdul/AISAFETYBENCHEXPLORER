## AISafetyBenchExplorer - Main Features

**AISafetyBenchExplorer** is a catalog of 180+ LLM Safety benchmarks, providing researchers with comprehensive metadata about available safety evaluation resources, their characteristics, accessibility, etc., which could be instrumental for conducting high-level statistical meta-analysis. 
The recorded features in the "Safety Evaluation Benchmarks" sheet include:

**Name** - The benchmark identifier or short name (e.g., "HarmEval", "ImplicitBias"). Serves as the unique label for each safety evaluation benchmark. Brief description of the benchmark provided on hover.

**Task Type** - Classification of the evaluation domain, including categories like "Safety", "Bias", "Jailbreak", "Red Teaming", "Content Moderation", "Medical Safety", "Fairness", and "Grounding". Benchmarks can be multi-labeled across multiple categories.

**Benchmark Paper Title** - The full title of the academic paper.

**Description**: One or two sentences introducing or describing the benchmark, providing the original research context and formal naming.

**Release** - The publication or release date of the benchmark (datetime format), indicating when the benchmark was made available to the research community.

**No. of Samples** - The quantitative size of the benchmark dataset, indicating how many test cases, prompts, or evaluation instances the benchmark contains (ranging from ~200 to 80,000+).

**Code** - The direct URL link to the source code repository (typically GitHub) where the benchmark implementation and associated code can be accessed.

**Dataset** - The direct URL link to where the benchmark dataset is hosted (typically Hugging Face, Kaggle, or direct GitHub repositories), or "not available" if not publicly shared.

**Created By** - The data creation methodology, categorized as "Human" (manual annotation), "Machine" (automated generation), or "Hybrid" (combination of both approaches).

**Entry Modalties** - The input format/type for benchmark entries, including "Prompts", "Conversations", "Examples", or other modality types that models are evaluated against.

**Dev Purpose** - The intended use case for the benchmark, such as "Eval only" (evaluation-focused) or "Eval and Train" (suitable for both evaluation and model training/fine-tuning).

**License** - The open-source license governing the benchmark's use and distribution (e.g., MIT, Apache-2.0, CC-BY-4.0, GPL-3.0), indicating usage rights and restrictions.

**Evaluation Metrics** - The specific quantitative and qualitative metrics used to assess model performance, such as "Attack Success Rate", "F1 Score", "Harmfulness Score", "Fairness Score", etc.

**Complexity level** - The difficulty or sophistication level of the benchmark, categorized as "Low", "Medium", or "High", reflecting evaluation challenge and nuance.

**Language Support** - The language(s) supported by the benchmark (predominantly "English" in this dataset, though may include multilingual support indicators).

**Integration Option** - The technical method for integrating the benchmark, typically "API" (programmatic access), "Export" (downloadable format), or other integration mechanisms.

**Citation Range** - The academic citation impact category, indicating citation frequency ranges such as "<50", "51-100", "101-500", suggesting research community adoption and influence.

**Cited by** - The exact number of academic citations received by the benchmark paper, providing a quantitative measure of research impact and adoption.

## Complexity Classifier


Benchmark complexity classification based on rigorous criteria:
**Popular**: Widely adopted benchmarks with high community impact. 
Criteria:
    - Citation count > 100 OR Frequent mention in safety literature
    - Used as baseline in multiple studies
    - Recognized standard in the field

**High**: Sophisticated evaluation requiring advanced reasoning. 
Criteria:
    - Multi-hop/compositional reasoning across >2 steps
    - Adversarial robustness testing (red-teaming, jailbreaking)
    - Subjective/open-ended generation tasks with nuanced evaluation
    - Risk-critical decision-making (medical, legal, financial safety)
    - Novel metric development or complex evaluation methodology
    - Requires domain expertise for annotation/evaluation

**Medium**: Moderate complexity with some challenging aspects. 
Criteria:
    - 1-2 step reasoning or limited compositional requirements
    - Some adversarial testing but not primary focus
    - Mix of objective and subjective evaluation
    - Standard metrics with minor adaptations
    - Moderate annotation/evaluation effort

**Low**: Straightforward evaluation with clear criteria. 
Criteria:
    - Single-step reasoning or direct lookup
    - Objective tasks with binary/categorical outcomes
    - Standard metrics (accuracy, F1) without modifications
    - Minimal adversarial considerations
    - Simple annotation process

**Classification Rules**:
- If benchmark exhibits Popular criteria, classify as Popular regardless of complexity
- For multi-faceted benchmarks, prioritize the most safety-critical dimension
- When in doubt between levels, choose higher complexity if adversarial/risk-critical
- Use 'Unknown' only when insufficient information is available

## Tools/Libraries & Flow Pipeline

```python
**Semantic Scholar API**
  https://www.semanticscholar.org/product/api
- Requires API Key request and approval
- Affiliation and purpose of use
- Prioritises research institutions, non-profit org, etc.
- Projected rate limit (1 request per second) / (100 requests/5mins - Public)
  **Provides**:
    - Authors, citations, venues, publication dates
    - Abstract, fields odf study
    - ArXiv/DOI/Pubmed IDs
    - Open Access status
    - Citation & reference counts
    - Influential citation metrics

**arXiv API**
  https://info.arxiv.org/help/api/index.html
- No API key request
- 1 request/3 seconds
  **Provides**:
    - Preprint Metadata
    - LaTex source access (when available)
    - Categories, comments
    - Direct PDF download links

**Unpaywall**
  https://unpaywall.org/products/api
- Email address required
- 100,000 requests/day 
- Polite crawling
  **Provides**:
    - Open Access Status
    - Legal PDF download links
    - OA version (published/accepted/submitted)
    - License information

**Crossref**
  https://www.crossref.org/documentation/retrieve-metadata/rest-api/
- No API key request
- Unlimited rate limit (Polite usage)
  **Provides**:
    - DOI resolution
    - Publisher metadata
    - References
    - Funder information

#===================================
# PARSERS
# ==================================

1. GROBID - XML Structure - <formula> tags for formula extraction - structure schema for sections/subsections - Cleaner text (runs on a server within a container)
2. Nougat - OCR - Document understanding Transformer architecture - Preserves LaTex math formula and complex equation - converts PDF to LateX format
3. Marker - OCR - open-source document parser - converts into markdown/json/chunks/html, mathematical equations to laTex format
4. PyMuPDF - OCR - text object - accesses title, author, apage count

Operations:
GROBID (primary)
    ↓ (fails)
Nougat (fallback 1)
    ↓ (fails)
Marker (fallback 2)
    ↓ (fails)
PyMuPDF (last resort)
```
